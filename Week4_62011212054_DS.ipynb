```python
import numpy as np
import math

def dist(a,b):
  return ( (a[0]-b[0])**2 + (a[1]-b[1])**2 )**(1/2)

def roundList(elist):
  for i in range(len(elist)):
    for j in range(len(elist[0])):
      elist[i,j] = round(elist[i,j],5)
  return elist

# DONE
def fuzzyCMeans(data):
  k1 = (6,7)
  k2 = (1,10)
  k3 = (7,5)
  k=3
  n = len(data)
  partition = np.random.rand(n,k)
  part_sum_row = partition.sum(axis=1)

  norm_part = partition/part_sum_row[:,np.newaxis]
  print(part_sum_row)
  
  norm_part_temp=[]

  iter = 150
  print("Start>>>")
  for cycle in range(iter):
    norm_part_temp = norm_part.copy()
    print("\n>>> Current Loop ",cycle,".")
    print("Patrition Matrix [ Before Update ]")
    print(norm_part)
    # calculate new k-means
    new_k=[]
    for j in range(k):
      ki1=0
      ki2=0
      sum_p=0
      for i in range(n):
        ki1 = ki1 + data[i][0] * (norm_part[i,j])**2
        ki2 = ki2 + data[i][1] * (norm_part[i,j])**2
        sum_p = sum_p+((norm_part[i,j])**2)
      ki1_new = ki1/sum_p
      ki2_new = ki2/sum_p
      new_k.append((ki1_new,ki2_new))

    print(".")
    print("New K-Means >>> ",new_k)

    # update partition
    for i in range(n):
      for j in range(k):
        x=dist(data[i],new_k[j])
        ki0=dist(data[i],new_k[0])
        ki1=dist(data[i],new_k[1])
        ki2=dist(data[i],new_k[2])
        norm_part[i,j] = 1/( (x/(ki0+ki1+ki2))**2 )

    # print("\SS>>>\n",norm_part)
    # print("\nTemp>>>\n",norm_part_temp)
    print("Patrition Matrix [ After Update ]")
    part_sum_row = norm_part.sum(axis=1)
    norm_part_new = norm_part/part_sum_row[:,np.newaxis]

    norm_part_temp = roundList(norm_part_temp)
    norm_part_new = roundList(norm_part_new)

    # print("temp = \n",norm_part_temp)
    # print(norm_part_new)

    isTrue = norm_part_new == norm_part_temp
    # print(isTrue)
    # print(isTrue.all())

    if isTrue.all():
      print(cycle,"________END of Cycle________")
      break
    else:
      # print("isTrue = ",isTrue)
      norm_part = norm_part_new
      print(norm_part)
      print("Continue>>")

  print("\nOutput:\nK-Means >> ",new_k)    
  print("Partition Matrix")
  print(norm_part_new)

data = [(7,8),(25,16),(13,2),(9,17),(1,1),(15,7),(8,1),(9,2),(11,7),(1,7)]

fuzzyCMeans(data)


```
