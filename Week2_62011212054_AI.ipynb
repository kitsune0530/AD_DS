```python
from collections import defaultdict

class Graph:
  def __init__(self):
    self.graph = defaultdict(list)
  def add_edge(self, u, v):
    self.graph[u].append(v)
  def show_graph(self):
    for n in self.graph:
      print(n,' -> ', self.graph[n])


graph = Graph()
graph.add_edge(1,2)
graph.add_edge(1,7)
graph.add_edge(1,8)
graph.add_edge(2,3)
graph.add_edge(2,6)
graph.add_edge(3,4)
graph.add_edge(3,5)
graph.add_edge(8,9)
graph.add_edge(8,12)
graph.add_edge(9,10)
graph.add_edge(9,11)

graph.show_graph()


```

```python
class Blindsearch(Graph):
  def __init__(self):
    super().__init__()
    self.visited = set()
    self.queue = []
  
  def clear(self):
    self.visited.clear()
    self.queue.clear()

  def DFS(self, v):
    print(v, end=' ')
    for n in self.graph[v]:
      if n not in self.visited:
        self.DFS(n)

  def BFS(self, v):
    self.visited.add(v)
    print(v, end=' ')
    for n in self.graph[v]:
      self.queue.append(n)
    for n in self.queue:
      if n not in self.visited:
        self.BFS(n)

# graph = Blindsearch()
# graph.add_edge(1,2)
# graph.add_edge(1,7)
# graph.add_edge(1,8)
# graph.add_edge(2,3)
# graph.add_edge(2,6)
# graph.add_edge(3,4)
# graph.add_edge(3,5)
# graph.add_edge(8,9)
# graph.add_edge(8,12)
# graph.add_edge(9,10)
# graph.add_edge(9,11)
# graph.BFS(1)

a = Blindsearch()
a.add_edge(9,10)
a.add_edge(9,11)
a.add_edge(1,2)
a.add_edge(1,7)
a.add_edge(1,8)
a.add_edge(2,3)
a.add_edge(2,6)
a.add_edge(3,4)
a.add_edge(3,5)
a.add_edge(8,9)
a.add_edge(8,12)
a.show_graph()
a.DFS(1)
print()
a.BFS(1)
```

```python
# Robot

text = open('/content/drive/MyDrive/Github/AI_DS/obstacle.txt', 'r')

class BlindsearchMatrix(Graph):
  def __init__(self):
    super().__init__()
    self.visited = set()
    self.queue = []
    self.added = []
    self.MSTQueue=[]
    self.x = 0
    self.y = 0
    self.DFSbreak = False
    self.distance = 0
  
  def clear(self):
    self.visited.clear()
    self.queue.clear()

# Done
  def MST(self, size,start,matrix):
    
    self.x = size[0]
    self.y = size[1]
    self.travel(matrix,start)
          
    # print(self.added)
      
  def travel(self, matrix, location):
    i = location[0]
    j = location[1]
    if (matrix[i][j]!=2):
        self.added.append([i,j])
        # print(i, ' ',j, end = ' ')
        if(j < self.y-1 and matrix[i][j+1]!=2 and [i,j+1] not in self.added):
            # print("right")
          self.add_edge(tuple([i,j]),tuple([i,j+1]))
          self.MSTQueue.append([i,j+1])
        if(i < self.x-1 and matrix[i+1][j]!=2 and [i+1,j] not in self.added):
           # print("down")
          self.add_edge(tuple([i,j]),tuple([i+1,j]))
          self.MSTQueue.append([i+1,j])
        if(j > 0 and matrix[i][j-1]!=2 and [i,j-1] not in self.added):
          # print("left")
          self.add_edge(tuple([i,j]),tuple([i,j-1]))
          self.MSTQueue.append([i,j-1])
        if(i > 0 and matrix[i-1][j]!=2 and [i-1,j] not in self.added):
            # print("up")
          self.add_edge(tuple([i,j]),tuple([i-1,j]))
          self.MSTQueue.append([i-1,j])
    for queue in self.MSTQueue:
      if ( queue not in self.added) :
        self.travel(matrix,queue)

  def DFS(self, v,stop):
    # print(v, end=' ')

    for n in self.graph[v]:
      self.distance=self.distance+1
      if not self.DFSbreak:
        if (n == stop):
          # print(n, end=' ')
          # print("END")
          print("distance = ",self.distance)
          self.DFSbreak = True
          break
        if n not in self.visited :
          self.DFS(n, stop)

  def BFS(self, v,stop):
    self.visited.add(v)
    print(v, end=' ')

    for n in self.graph[v]:
      self.queue.append(n)
    for n in self.queue:
      if (tuple(v) == tuple(stop)):
        print(n, end=' ')
        print("END")
      if n not in self.visited:
        self.BFS(n,stop)

lines = []
with text as f:
  lines = f.readlines()


size = [int(x) for x in lines[0].split()]
print('Size = ',size)

start = [int(x) for x in lines[int(size[0])+1].split()]
stop = [int(x) for x in lines[int(size[0])+2].split()]
print('Start = ',start)
print('Stop = ',stop)

graph = lines[1:int(size[0])+1]
matrix = []
for i in graph:
  temp=[int(x) for x in i.split()]
  matrix.append(temp)

# print(matrix)
a = BlindsearchMatrix()
a.MST(size,start,matrix)
a.show_graph()
a.DFS(tuple([0,0]), tuple(stop))
  
text.close()
```
