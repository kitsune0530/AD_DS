```python
#17
netAmount = 0
while True:
  s = input()
  if not s:
    break
  values = s.split(" ")
  operation = values[0]
  amount = int(values[1])
  if operation=="D":
    print("+")
    netAmount+=amount
  elif operation=="W":
    print("-")
    netAmount-=amount
  else:
    pass
print(netAmount)
```

```python
#18
import re
value = []
items = [x for x in input().split(',')]
for p in items:
  if len(p) < 6 or len(p)>12:
    continue
  else:
    pass
  # print(p)
  if not re.search("[a-z]",p):
    continue
  elif not re.search("[0-9]",p):
    continue
  elif not re.search("[A-Z]",p):
    continue
  elif not re.search("[$#@]",p):
    continue
  elif re.search("\s",p):
    continue
  else:
    pass
  # print("debug")
  value.append(p)

print(",".join(value))


```

```python
#20
def putNumbers(n):
  i = 0
  while i<n:
    j=i
    i=i+1
    if j%7==0:
      yield j
      
for i in reversed(list(putNumbers(100))):
  print(i)

```

```python
#21
import math
pos = [0,0]
while True:
  s = input()
  if not s:
    break
  movement = s.split(" ")
  direction = movement[0]
  steps = int(movement[1])
  if direction == "UP":
    pos[0]+=steps
  elif direction == "DOWN":
    pos[0]-=steps
  elif direction == "LEFT":
    pos[1]+=steps
  elif direction == "RIGHT":
    pos[1]-=steps
  else:
    pass
  
print(int(round(math.sqrt(pos[1]**2+pos[0]**2))))
```
