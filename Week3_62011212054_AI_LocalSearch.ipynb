```python
import random

def findWeight(tsp, solution):
  length = len(tsp)
  weight = 0
  for i in range(length):
    if i < len(tsp)-1:
      weight = weight+tsp[solution[i]][solution[i+1]]
      # print("plus",tsp[solution[i]][solution[i+1]])
    else:
      weight = weight+tsp[solution[i]][solution[0]]
      # print("plus",tsp[solution[i]][solution[0]])


  return weight

def TSP(tsp):
  length = len(tsp)
  solution = [i for i in range(0,length)]
  random.shuffle(solution)
  print("Current Solution:",solution)

  while True:
    weight = findWeight(tsp,solution)
    print("Current Weight:",weight)

    possibleSolution = []
    possibleWeight = []
    # Swap City
    for i in range(length):
      tempSolution = solution[:]
      temp = tempSolution[i]
      if (i<length-1):

        tempSolution[i] = tempSolution[i+1]
        tempSolution[i+1] = temp
      else:
        tempSolution[i] = tempSolution[0]
        tempSolution[0] = temp

      possibleSolution.append(tempSolution)
      possibleWeight.append(findWeight(tsp,possibleSolution[i]))

    for i in range(length):
      print("Possible ",i+1,":",possibleSolution[i],'=',possibleWeight[i])

    new_weight = min(possibleWeight)
    new_weight_index = possibleWeight.index(new_weight)

    if (new_weight>=weight):
      print("___________________________")
      print("\n>>Output\nBest Solution = ",solution,"\nWeight = ",weight)
      break
    else:

      solution = possibleSolution[new_weight_index]
      print("Better Solution >>",solution,"Weight = ",new_weight," index: ", new_weight_index)
      print()


tsp = [[0,100,550,300],
       [100,0,300,500],
       [550,300,0,100],
       [300,500,100,0]]
       
tsp2 = [[0,100,550,300,150,50],
       [100,0,300,500,300,30],
       [550,300,0,100,100,10],
       [300,500,100,0,120,170],
       [150,300,100,120,0,200],
       [50,30,10,170,200,0]]

TSP(tsp2)










```

```python
# https://www.geeksforgeeks.org/traveling-salesman-problem-tsp-implementation/

from sys import maxsize
from itertools import permutations
V = 6
 
# implementation of traveling Salesman Problem
def travellingSalesmanProblem(graph, s):
 
    # store all vertex apart from source vertex
    vertex = []
    for i in range(V):
        if i != s:
            vertex.append(i)
 
    # store minimum weight Hamiltonian Cycle
    min_path = maxsize
    next_permutation=permutations(vertex)
    for i in next_permutation:
 
        # store current Path weight(cost)
        current_pathweight = 0
 
        # compute current path weight
        k = s
        for j in i:
            current_pathweight += graph[k][j]
            k = j
        current_pathweight += graph[k][s]
 
        # update minimum
        min_path = min(min_path, current_pathweight)
         
    return min_path
 
 
# Driver Code
if __name__ == "__main__":
 
    # matrix representation of graph

    tsp2 = [[0,100,550,300,150,50],
          [100,0,300,500,300,30],
          [550,300,0,100,100,10],
          [300,500,100,0,120,170],
          [150,300,100,120,0,200],
          [50,30,10,170,200,0]]
    s = 0
    print(travellingSalesmanProblem(tsp2, s))
```
