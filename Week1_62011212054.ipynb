```python
# Lab 1-1

def fibonacciRecur(x):
  if (x<=0) :
    return 0 
  if x==1:
    return 1
  else :
    return fibonacciRecur(x-1)+fibonacciRecur(x-2)


def fibonacciIter(x):
  f1=0
  f2=1
  t=f1+f2
  if x == 0:
    print(0)
  if x==1:
    print(1)
  if x==2:
    print(2)
  
  for i in range(3,x+1):
    f1=f2
    f2=t
    t=f1+f2
  print("Iterative = ",t)

print("Please Input Your Number (Positive Number)");
x = int(input())
fibonacciIter(x)

print("Recursive = " , fibonacciRecur(x))
```

```python
# Lab 1-2

import numpy as np
def LCS(x,y):
  n1 = len(x)
  n2 = len(y)
  c = np.zeros((n1+1,n2+1))

  # print(c)
  for i in range(0,n1):
    for j in range(0,n2):

      if x[i] == y[j]:
        c[i+1,j+1] = c[i,j]+1
      else:
        c[i+1,j+1] = max(c[i,j+1],c[i+1,j])
  return c

# Done
def backtrack(c, x, y, i, j):
  print("Current at",i," ",j)
  if i==0 or j==0:
    return ""
  if x[i-1] == y[j-1]:
    return backtrack(c,x,y,i,j-1)+x[i-1]
  if c[i,j-1] > c[i-1,j]:
    return backtrack(c,x,y,i,j-1)
  # if c[i,j] > c[i-1,j] and c[i,j] > c[i,j-1]:
  #   print(" Check")
  #   return backtrack(c,x,y,i,j-1)+x[i-1]
  return backtrack(c,x,y,i-1,j)

seq1 = 'XMJYAUZ'
seq2 = 'MZJAWXU'
# seq1 = 'abcab'
# seq2 = 'aabaca'
len1 = len(seq1)
len2 = len(seq2)

c = LCS(seq1,seq2)
# print(c)
print(backtrack(c,seq1,seq2,len1,len2))
print(c)
```
