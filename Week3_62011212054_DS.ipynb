```python
from sklearn.cluster import KMeans
# ก้อน Data 2D
from sklearn.datasets import make_blobs, make_moons
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# ตัวอย่าง 500 ค่ากลาง 5 การกระจาย 0.6
# x,y = make_blobs(n_samples=500,centers = 5, cluster_std=0.9, random_state = 0)
x,y = make_moons(n_samples=500,noise=0.15,random_state=5)
# print(x,y)

plt.scatter(x[:,0],x[:,1],s=5,c='red')
plt.show()

# paremeter(จำนวน cluster)
# kmeans = KMeans(n_clusters=5)
kmeans = KMeans(n_clusters=2)
# นำ Data มาจัดกลุ่ม
kmeans.fit(x)
y_predict = kmeans.predict(x)
print(y)
print(y_predict)
plt.scatter(x[:,0],x[:,1],s=5,c=y_predict)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
# facny plot
import plotly.offline as ply
import plotly.figure_factory as ff

data = pd.read_csv('/content/drive/MyDrive/Github/AI_DS/DS_data/Mall_Customers.csv')

# data.head()

# table to show data
dat = ff.create_table(data.head(10))
ply.iplot(dat)



```

```python
# make sure data has no missing data ( 2D )
print(data.isnull().any().any())

#summary
# print(data.describe())

desk = ff.create_table(data.describe())
ply.iplot(desk)


```

```python
# try plot distribution
import warnings
warnings.filterwarnings('ignore')
plt.rcParams["figure.figsize"] = (10,7)

plt.subplot(121)
sb.set_style(style="whitegrid")
sb.distplot(data['Annual Income (k$)'],color='red')
plt.title("Distribution of Income")
plt.xlabel('range of income')
plt.ylabel('count')

plt.subplot(122)
sb.distplot(data['Age'],color='blue')
plt.title("Distribution of Age")
plt.xlabel('age')
plt.ylabel('count')
plt.show()
```

```python
plt.rcParams["figure.figsize"]=20,7

# bar graphs 
sb.countplot(data['Annual Income (k$)'],palette='rainbow')
plt.show()
```

```python
# correlation ความสัมพันธ์ของ data ในแต่ละ D
final_data = data.drop('CustomerID',axis=1)
final_data.corr()

plt.rcParams['figure.figsize']=4,4
sb.heatmap(final_data.corr())
plt.show()
```

```python
from mpl_toolkits import mplot3d

# ทุก row เอา col 2 , 4
# .values เพื่อตัด Head
# แยกกลุ่มที่มีกำลังจ่าย
# x = data.iloc[:,[2,3,4]].values
x = data.iloc[:,[2,3,4]].values
print(x.shape)


plt.scatter(x[:,0],x[:,1],s=20,c='g')
plt.show()

kmeans = KMeans(4)
kmeans.fit(x)
y_predict = kmeans.predict(x)

plt.scatter(x[:,0],x[:,1],s=20,c=y_predict)
plt.scatter
plt.show()



fig = plt.figure(figsize = (10, 7))
ax = plt.axes(projection ="3d")
# Creating plot
ax.scatter3D(x[:,0],x[:,1],x[:,2] ,color = "green")
plt.title("simple 3D scatter plot")
# show plot
plt.show()


# x.head(10)
```
