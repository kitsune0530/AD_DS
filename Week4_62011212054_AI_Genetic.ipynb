```python
# ทำ Crossover
import numpy as np
def objective(x):
  return x[0]**3 + x[1]**2 - (4*x[2])

def decode(pop,n_bit,bounds):
  res = []
  largest = 2**(n_bit)
  # print(pop)
  temp = pop[0:10]
  str_bit1 = ''.join(str(i) for i in temp)
  val1 = int(str_bit1,2)
  temp = pop[10:20]
  str_bit2 = ''.join(str(i) for i in temp)
  val2 = int(str_bit2,2)
  temp = pop[20:30]
  str_bit3 = ''.join(str(i) for i in temp)
  val3 = int(str_bit3,2)
  # print(val1,' ',val2,' ',val3)

  norm1 = bounds[0][0]+(val1/largest)*(bounds[0][1]-bounds[0][0])
  norm2 = bounds[1][0]+(val2/largest)*(bounds[1][1]-bounds[1][0])
  norm3 = bounds[2][0]+(val3/largest)*(bounds[2][1]-bounds[2][0])
  res = [norm1,norm2,norm3]
  return res

def selection(population,scores): 
  # สุ่มมา 3-4 ตัว
  # rand ตัวเริ่มต้นเพื่อเปรียบเทียบ
  c_idx = np.random.randint(len(population))
  for ix in np.random.randint(0,len(population),4):
    if scores[ix] < scores[c_idx]:
      c_idx = ix;
  return pop[c_idx]

def cross_over(p1,p2,t):
  c1,c2 = p1.copy(),p2.copy()
  r = np.random.rand()
  # print(r)
  if r>t:
    pt = np.random.randint(0,len(p1)-2)
    # c1 = p2[pt:]+p1[:pt]
    c1 = p1[:pt]+p2[pt:]
    c2 = p1[pt:]+p2[:pt]
    # print(pt)
  return [c1,c2]

def mutation(c,t):
  for i in range(len(c)):
    if(np.random.rand()<t):
      c[i] = 1-c[i]

n_bit = 10
n_pop = 100
#กำหนดรอบ
n_iter = 50

# กำหนดช่วงข้อมูล
bounds = [[-5,5],[-10,10],[-5,5]]
# สุ่ม population
pop = [np.random.randint(0,2,n_bit*3).tolist() for i in range(0,n_pop)]
# decode binary to decimal
decodeds = [decode(p,n_bit,bounds) for p in pop]
# objective function
scores = [objective(d) for d in decodeds]

# ค่าที่ดีที่สุด
best = 0
best_pop = pop[0]


# เลือก parent


for k in range(0,n_iter):
  decodeds = [decode(p,n_bit,bounds) for p in pop]
  score = [objective(d) for d in decodeds]
  for i in range(n_pop):
    if(score[i]<best):
      best = score[i]
      best_pop = pop[i]
  print("score:",best,' pop:',best_pop," :",decode(best_pop, n_bit, bounds))
  parent = [selection(pop,scores) for i in range(n_pop)]
  child = []
  for i in range (0,n_pop, 2):
    p1,p2 = parent[i],parent[i+1]
    [c1,c2] = cross_over(p1,p2,0.7)
    mutation(c1,0.8)
    mutation(c2,0.8)
    child.append(c1)
    child.append(c2)
  pop = child




# ทำ cross-over ทีละคู่เพื่อสร้างลูกขึ้นมาใหม่ ไม่จำเป็นต้องทำทั้งหมด
[c1, c2] = cross_over(parent[0],parent[1],-1)

print(parent[0],' ',parent[1])
print(c1,' ',c2)


for p in scores:
  pass
  # print(p)
for p in parent:
  pass
  # print(p)
for c in child:
  pass
  # print(c)

print(len(parent))
```
