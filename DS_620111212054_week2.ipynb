```python
data = pd.read_csv('/content/drive/MyDrive/Github/AI_DS/DS_data/BL-Flickr-Images-Book.csv', encoding="ISO-8859-1")
# data.head()


# ทำการ cleaning โดยการ drop col
li_to_drop = ['Edition Statement', 'Corporate Author', 'Corporate Contributors', 'Former owner', 'Engraver']

# axis 0 = row, 1 = col และ inplace คือการลบภายในไฟล์เลย
# data.drop(li_to_drop, inplace=True, axis=1)
data.drop(columns=li_to_drop, inplace=True)
data.head()

# check if column is unique
print(data['Identifier'].is_unique)

#set index to col 'Identifier'
data = data.set_index('Identifier')
data.head(20)

# data.iloc[5]
# print(data.loc[216])
# print(data.loc[216,'Title'])


# plt.figure(figsize=(16,6))
# sb.lineplot(data = data)
```

```python
# Cleaning specific column by using Regula Expression
data.loc[1905:,'Date of Publication'].head()

# ^ = group
# \d Decimal
# {4} 4 digits
regEx = r'^(\d{4})'

# เลือกข้อมูลจาก col แปลงให้เป็น String แล้วทำการ Extract ด้วย regExpress โดย return as Series ยังไม่ได้เปลี่ยนใน Data
extract = data['Date of Publication'].str.extract(regEx, expand = False)
extract.head(100)
```

```python
li_place = ['London','Oxford']
pub = data['Place of Publication']

# ถ้ามีคำว่า
london = pub.str.contains('London')
oxford = pub.str.contains('Oxford')

series = pd.DataFrame([london,oxford])
print(series)

# clean ค่าให้เหลือเฉพาะที่มีคำว่า London และ Oxford
data['Place of Publication'] = np.where(london, 'London', np.where(oxford, 'Oxford', pub.str.replace('-','')))
data['Place of Publication'].head(50)


```
